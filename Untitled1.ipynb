```python
df.info(memory_usage='deep')
```

```python
for dtype in ['float64', 'int64', 'object']:
    selected_dtypes = df.select_dtypes(include=dtype)
    mean_usage = selected_dtypes.memory_usage(deep=True).mean()
    print(f'Average usage for {dtype}: {mean_usage} bytes')
```

```python
df_float = df.select_dtypes(include='float').fillna(0)
converted_float = df_float.astype(int)

convert = lambda a: f'{(a/1024 ** 2):03.2f} MB'

print(convert(df_float.memory_usage(deep=True).sum()))
print(convert(converted_float.memory_usage(deep=True).sum()))
```

```python
compare_types = pd.concat([df_float.dtypes, converted_float.dtypes], axis=1)
compare_types.columns = ['Було', 'Стало']
compare_types.apply(pd.Series.value_counts)
```
